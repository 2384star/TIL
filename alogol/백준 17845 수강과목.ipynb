```python
import sys
input= sys.stdin.readline
import torch
```

```python
N, K = map(int, input().rstrip().split())
```

```python
DP=[[0 for _ in range(N+1)] for _ in range(K+1)] # N+1 K+1차원으로 0 초기화
value= [0]
time= [0]

for _ in range(K):
    v, t = map(int, input().rstrip().split())
    value.append(v)
    time.append(t)
```

```python
for i in range(1,K+1):#각 수강과목에 따른 반복 
    for j in range(1,N+1):
        if time[i]>j: #만약 수강시간이 한도보다 크다면 이전 수강시간까지 고려햇을때의 최대가치를 사용한다
            DP[i][j]= DP[i-1][j]
        else:
            DP[i][j] = max(DP[i-1][j - time[i]] + value[i], DP[i-1][j]) #아니라면? 
            #즉 잘생각을해보자 time[2]=50이라고치면 우리는 j가 51이 될때 최대한도(90-50)을 수행할것이다 
            #그러면 40이 남을태고 즉 우리는 time[2]에 대한 한도를 처리한것이다 그후 [1][40]이 시행될태고 time[1]의값이 40미만이엿다면 그값이 더해질것이다 
            #전에 가방에 넣던 점수에 현재  점수를추가하고 점수를 추가하지 않앗을때의 점수를 비교해서 큰값을 구한다 
    
print(DP[K][N])
```

# 모를때정리
- i=2, 즉 물건 2를 고려할 때, 이 물건의 무게(time[2])는 50입니다.
j=51, 즉 무게 한도가 51일 때, 물건 2를 배낭에 넣을 수 있습니다.
물건 2를 배낭에 넣었을 때 남는 무게 한도는 51 - time[2] = 51 - 50 = 1입니다.
DP[1][1]을 조회합니다. 이 값은 물건 1만 고려했을 때 무게 한도 1에서의 최대 가치입니다.
DP[2][51]은 물건 2를 넣었을 때의 가치와 DP[1][1]을 더한 값과, 물건 2를 넣지 않았을 때의 가치인 DP[1][51] 중 더 큰 값을 저장합니다.

```python
N = 5  # 무게 한도
K = 3  # 물건의 개수
value = [0, 1, 4, 5]  # 각 물건의 가치 (0번 인덱스는 사용하지 않음)
time = [0, 1, 2, 3]   # 각 물건의 무게 (0번 인덱스는 사용하지 않음)

# DP 테이블 초기화:
DP = [[0 for _ in range(N+1)] for _ in range(K+1)]

# 초기 DP 테이블 상태 출력
print("Initial DP table:")
for row in DP:
    print(row)
print("\n")
```
