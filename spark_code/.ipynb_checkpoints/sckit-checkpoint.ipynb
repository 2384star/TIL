```python
import os
import pandas as pd

folder = "dataset/TEST_AWS"

# test_aws 폴더 내 모든 CSV 파일 목록을 가져옵니다.
csv_files = [f for f in os.listdir(folder) if f.endswith('.csv')]

# 각 CSV 파일을 불러와서 하나의 리스트에 저장합니다.
dataframes = []

for file in csv_files:
    file_path = os.path.join(folder, file)
    df = pd.read_csv(file_path)
    dataframes.append(df)

# 모든 데이터프레임을 하나로 합칩니다.
combined_df = pd.concat(dataframes, ignore_index=True)
```

```python
# 일시 칼럼의 데이터를 '월-일 시:분' 형식의 문자열로 변환
combined_df['일시'] = pd.to_datetime(combined_df['일시'], format="%m-%d %H:%M")

# 월, 일, 시간, 분을 추출하여 별도의 칼럼으로 저장
combined_df['월'] = combined_df['일시'].apply(lambda x: x.month)
combined_df['일'] = combined_df['일시'].apply(lambda x: x.day)
combined_df['시간'] = combined_df['일시'].apply(lambda x: x.hour)
combined_df['분'] = combined_df['일시'].apply(lambda x: x.minute)

# 지점 열을 원-핫 인코딩으로 변환
combined_df = pd.concat([combined_df, pd.get_dummies(combined_df['지점'], prefix='지점')], axis=1)
combined_df = combined_df.drop(['지점'], axis=1)

# 회귀 분석에 사용할 변수 목록
variables_to_fill = ['기온(°C)', '풍향(deg)', '풍속(m/s)', '강수량(mm)', '습도(%)']

for var in variables_to_fill:
    # 결측치가 없는 행과 열을 선택
    not_null_rows = combined_df[var].notnull()
    not_null_df = combined_df.loc[not_null_rows].drop(['일시'], axis=1)

    # 회귀 분석을 수행할 X, y 데이터 생성
    X = not_null_df.drop(var, axis=1)
    y = not_null_df[var]

    # 회귀 분석 수행
    model = LinearRegression()
    model.fit(X, y)

    # 결측치가 있는 행의 인덱스를 가져옴
    null_rows = combined_df[var].isnull()
    null_df = combined_df.loc[null_rows].drop(['일시'], axis=1)

    # 결측치를 채울 데이터 준비
    X_to_fill = null_df.drop(var, axis=1)

    # 결측치를 채워 넣기
    combined_df.loc[null_rows, var] = model.predict(X_to_fill)

# 결측치가 채워진 데이터프레임을 CSV 파일로 저장
combined_df.to_csv("backup_data.csv", index=False)
```
