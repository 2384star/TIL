### 우선 선형 모델을 골라서 시도해보자
- t_c= t_u * w +b

```python
def model(t_u,w,b): # w:가중치 b:편향
    return w* t_u +b
```

- mse: 평군 제곱손실 
- 오차범위를 제곱하여 평균내는것

```python
def loss_fn(t_p, t_c):
    squared_diffs= (t_p -t_c)**2
    return squared_diffs.mean()
```

```python
dalta=0.1

loss_rate_of_change_w= \
(loss_fn(model(t_u,w+dalta,b),t_c) -
loss_fn(model(t_u,w-dalta,b),t_c))/(2.0*dalta)
```

```python
#임의의 학습률 지정 
learning_rate= 1e-2

w= w- learning_rate *loss_rate_of_change_w
```

## 미분식
d_loss_fn / d_w=(d_loss_fn/d_t_p)*(d_t_p /d_w)

```python
def loss_fn(t_p,t_c):
    squared_diffs=(t_p-t_c)**2
    return squared_diffs.mean()
```

```python
def dloss_fn(t_p,t_c):
    dsq_diffs = 2*(t_p -t_c) /t_p.size(0)
    return dsq_diffs
```

```python
def model(t_u,w,b):
    return w* t_u +b 
```

```python
def dmodel_dw(t_u,w,b):
    return t_u
    
```

```python
def dmodel_db(t_u,w,b):
    return 1.0
```

- 경사함수 정의

```python
def grad_fn(t_u,t_c,t_p,w,b):
    dloss_dtp=dloss_fn(t_p,t_c)
    dloss_dw=dloss_dtp *dmodel_dw(t_u,w,b)
    dloss_db=dloss_dtp * dmodel_db(t_u,w,b)
    
    return torch.stack([dloss_dw.sum(),dloss_db.sum()]) 
```

### 모델의 적합도 판단
- 에포크 : 훈련 셈플을 가지고 반복적으로 파라미터를 조정하는 훈련의 한단위

```python
def training_loop(n_epochs,learning_rate,params,t_u,t_c):
    for epoch in range(1,n_epochs+1):
        w,b=params
        t_p=model(t_u,w,b)
        loss=loss_fn(t_p,t_c)
        grad=grad_fn(t_u,t_c,t_p,w,b) #역방향
        
        params=params -learning_rate *grad
        print("Epoch %d Loss %f " % (epoch,float(loss)))
```

```python
#루프실행
training_loop(1000,1e-2,torch.tensor([1.0,2.0]),t_u,t_c)
```

```python
def training_loop(n_epochs, learning_rate, params, t_u, t_c,
                  print_params=True):
    for epoch in range(1, n_epochs + 1):
        w, b = params

        t_p = model(t_u, w, b)  # <1>
        loss = loss_fn(t_p, t_c)
        grad = grad_fn(t_u, t_c, t_p, w, b)  # <2>

        params = params - learning_rate * grad

        if epoch in {1, 2, 3, 10, 11, 99, 100, 4000, 5000,10000}:  # <3>
            print('Epoch %d, Loss %f' % (epoch, float(loss)))
            if print_params:
                print('    Params:', params)
                print('    Grad:  ', grad)
        if epoch in {4, 12, 101}:
            print('...')

        if not torch.isfinite(loss).all():
            break  # <3>
            
    return params
```

```python
#루프실행
training_loop(100,1e-2,torch.tensor([1.0,0.0]),t_u,t_c)
```

### 주의 backward 호출은 미분을 말단 노드에 누적시킨다 즉 저장이아님

- 아래코드는 예시임

```python
if params.grad is not None:
    params.grad.zero_()
```

```python
import torch.nn as nn
import torch.optim as optim

# 간단한 신경망 정의
model = nn.Sequential(
    nn.Linear(10, 50),  # 입력층
    nn.ReLU(),
    nn.Linear(50, 30),  # 히든 층
    nn.ReLU(),
    nn.Linear(30, 1)   # 출력층
)

optimizer = optim.SGD(model.parameters(), lr=0.01)

# 임의의 입력과 정답 데이터
input_data = torch.randn(64, 10)
labels = torch.randn(64, 1)

# 손실 함수 정의
loss_fn = nn.MSELoss()

# 훈련 반복
for epoch in range(100):
    optimizer.zero_grad()  # <---- 여기서 모든 기울기를 0으로 초기화합니다.
    
    outputs = model(input_data)
    loss = loss_fn(outputs, labels)
    
    loss.backward()  # <---- 이곳에서 전체 네트워크의 역전파가 이루어집니다.
    optimizer.step()  # <---- 모델의 가중치 업데이트
```

```python
#1. 모델갖고오기

def model(t_u,w,b): # w:가중치 b:편향
    return w* t_u +b

def loss_fn(t_p, t_c):
    squared_diffs= (t_p -t_c)**2
    return squared_diffs.mean()
```

## 순서 

순전파: t_p = model(t_u, *params)를 통해 모델을 사용하여 예측값을 계산합니다.

손실 계산: loss = loss_fn(t_p, t_c)를 통해 예측값과 실제값 사이의 손실을 계산합니다.

역전파: loss.backward()를 통해 손실에 대한 그래디언트를 계산합니다. 이 때, 모델의 파라미터들에 대한 그래디언트 값이 params.grad에 저장됩니다.

파라미터 업데이트: with torch.no_grad(): 내에서 그래디언트 디센트를 사용하여 파라미터를 업데이트합니다.

로그 출력: 특정 에폭마다 손실값을 출력합니다.

```python
def training_loop(n_epochs, learning_rate, params,t_u,t_c):
    for epoch in range(1,n_epochs+1):
        if params.grad is not None:
            params.grad.zero_() 
        
        t_p=model(t_u, *params)
        loss=loss_fn(t_p,t_c)
        loss.backward()
        #순전파 부분 
        with torch.no_grad():
            params -=learning_rate * params.grad
        #torch.no_grad() 자동미분 계산 일시 비활성화
        
## 500텀 마다 출력            
        if epoch % 500 ==0:
            print("Epoch %d, Loss %f " %(epoch,float(loss)))
    
    return params
```

```python
training_loop(
    n_epochs=5000,
    learning_rate=1e-2,
    params=torch.tensor([1.0,0.0],requires_grad=True),
    t_u=t_un,
    t_c=t_c)
```

```python
def training_loop(n_epochs, optimizer, params,t_u,t_c):
    for epoch in range(1,n_epochs+1):
        if params.grad is not None:
            params.grad.zero_() 
        
        t_p=model(t_u, *params)
        loss=loss_fn(t_p,t_c)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        
## 500텀 마다 출력            
        if epoch % 500 ==0:
            print("Epoch %d, Loss %f " %(epoch,float(loss)))
    
    return params
```

```python
params=torch.tensor([1.0,0.0],requires_grad=True)
learning_rate=1e-2
optimizer=optim.SGD([params],lr=learning_rate)

training_loop(
n_epochs=5000,
optimizer=optimizer,
params=params,
t_u=t_un,
t_c=t_c)
```
