```python
# 각 컬럼의 value_counts() 결과를 세분화하여 저장하고 CSV로 내보내는 코드를 작성합니다.

# 컬럼별로 value_counts() 결과를 세분화하여 저장할 빈 리스트 생성
value_counts_list = []

# 각 컬럼에 대해 value_counts를 계산합니다.
for col in data2.columns:
    # 각 값과 그 값이 나타난 횟수를 가져옵니다.
    counts = data2[col].value_counts().to_dict()
    # 컬럼 이름과 함께 리스트에 저장합니다.
    for value, count in counts.items():
        value_counts_list.append({'Column': col, 'Value': value, 'Count': count})

# 세분화된 결과를 데이터프레임으로 변환
value_counts_df = pd.DataFrame(value_counts_list)

# 결과 데이터프레임을 CSV 파일로 저장합니다.
value_counts_csv_path = './eda2.csv'
value_counts_df.to_csv(value_counts_csv_path, index=False)

value_counts_csv_path  # 저장된 파일의 경로를 반환합니다.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
# 서브플롯 설정
num_columns = len(data2.columns)
n_rows = int(np.ceil(num_columns / 4))  # 4열로 서브플롯을 배치
fig, axes = plt.subplots(n_rows, 4, figsize=(20, n_rows * 5))

# 모든 컬럼에 대해 반복
for i, column in enumerate(data4.columns):
    # 각 값의 빈도 계산
    counts = data4[column].value_counts()
    # 빈도를 y축 값으로 사용하여 산점도 데이터 준비
    scatter_data = pd.DataFrame({
        'Value': np.repeat(counts.index, counts.values),
        'Frequency': np.concatenate([np.arange(c) for c in counts.values])
    })
    # 현재 컬럼의 서브플롯 위치 계산
    ax = axes[i // 4, i % 4]
    # 산점도 그리기
    sns.scatterplot(data=scatter_data, x='Value', y='Frequency', ax=ax)
    ax.set_title(column)
    ax.set_xlabel('value')
    ax.set_ylabel('count')

# 남은 서브플롯 숨기기
for j in range(i + 1, n_rows * 4):
    axes[j // 4, j % 4].set_visible(False)

plt.tight_layout()
plt.show()
```

### 전처리_CH3COOH_std': 1.0 면  '배합_빙초산_std': 0.2, 를가진다

- 전처리_CH3COOH"라는 용어는 염색 공정과 관련된 전처리 단계에 사용되는 아세트산(CH3COOH)을 나타낼 수 있습니다. 아세트산은 흔히 식초의 주요 성분으로 알려져 있지만 화학 제조 및 산업 공정에서도 사용되는 유기 화합물입니다.

염색 공정에서의 "전처리"는 섬유나 직물을 염색하기 전에 수행하는 여러 가지 처리 과정을 의미할 수 있습니다. 아세트산은 이런 전처리 과정 중 하나로 사용될 수 있는데, 이는 섬유의 pH를 조절하거나 특정 화학 반응을 촉진시키기 위해 산성 조건을 만들어주는 역할을 할 수 있습니다. 예를 들어, 아세트산은 양모나 나일론 같은 단백질 섬유나 합성 섬유의 염색을 준비하기 위해 사용될 수 있으며, 염료와 섬유 사이의 결합을 돕거나, 염색 과정에서 필요한 다른 화학 물질의 효과를 향상시키기 위해 섬유의 표면을 처리하는 데 사용될 수 있습니다.

- "배합 빙초산"이란 말은 일반적으로 '아세트산'을 지칭합니다. 빙초산은 식초의 주성분으로, 식초에 신맛을 내는 용도로 사용됩니다. 식용으로 안전한 빙초산은 식품첨가물로 사용되며, 식초에 3~5% 정도 함유되어 있습니다

### 사용이유
pH 조절: 염료의 효과적인 부착을 위해서는 종종 염색 욕의 적절한 pH 수준을 유지하는 것이 중요합니다. 아세트산은 약산으로서, 염색 용액의 pH를 조절하여 염료와 섬유 사이의 최적의 상호작용을 이끌어낼 수 있습니다.

염색 고착: 특정 염료, 특히 산성 염료와 반응성 염료는 산성 조건에서 섬유에 더 잘 고착됩니다. 아세트산을 사용하여 산성 조건을 만들어 염료가 섬유에 더 잘 결합하도록 돕습니다.

부식 방지: 빙초산은 섬유를 부식시키지 않으면서 산성 조건을 제공할 수 있습니다. 이는 염색 기계나 용기에 부식을 줄일 수 있는 안전한 선택을 제공합니다.

염료의 용해도 증가: 어떤 염료들은 약간의 산성 조건에서 용해도가 증가합니다. 이는 염료의 균일한 분포와 섬유에 대한 침투를 향상시킬 수 있습니다.

```python
# Let's update the code to include the unique value counts for both dataframes std1 and std2, along with the value names, in a single CSV file.

def create_unique_value_counts_csv(df1, df2):
    # Calculate the value counts for each column in the dataframes
    value_counts_df1 = {column: df1[column].value_counts() for column in df1.columns}
    value_counts_df2 = {column: df2[column].value_counts() for column in df2.columns}

    # Filter out the columns that have only one unique value and format the results
    single_value_columns_df1 = {column: values.index[0] for column, values in value_counts_df1.items() if len(values) == 1}
    single_value_columns_df2 = {column: values.index[0] for column, values in value_counts_df2.items() if len(values) == 1}

    # Create a combined list of dictionaries for each dataframe
    combined_list = [
        {"DataFrame": "std1", "Column": column, "Unique Value": value, "Count": df1[column].value_counts().iloc[0]}
        for column, value in single_value_columns_df1.items()
    ] + [
        {"DataFrame": "std2", "Column": column, "Unique Value": value, "Count": df2[column].value_counts().iloc[0]}
        for column, value in single_value_columns_df2.items()
    ]

    # Convert the combined list to a DataFrame
    combined_df = pd.DataFrame(combined_list)

    # Export the DataFrame to a CSV file
    csv_file_path = './eda5.csv'
    combined_df.to_csv(csv_file_path, index=False)

    return combined_df, csv_file_path

# Create the CSV file
unique_value_counts_df, csv_file_path = create_unique_value_counts_csv(std1, std2)

# Display the dataframe and the path to the CSV file
unique_value_counts_df, csv_file_path
```
