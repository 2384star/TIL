```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("전처리저장용.csv")
df_cate= df[df['카테고리'].isin(['데이터 사이언스', '일러스트'])]
```

```python
df_grouped =df_cate.groupby('월')['카테고리'].value_counts()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] ='AppleGothic'
# 그래프 크기 설정
plt.figure(figsize=(12, 12))
df_grouped_unstacked = df_grouped.unstack(level=-1)
df_grouped_unstacked.plot(kind='line')
# x축의 눈금을 1 단위로 설정
plt.xticks(range(min(df_grouped_unstacked.index), max(df_grouped_unstacked.index)+1))

plt.title('월별비교')
plt.xlabel('판매량')
plt.ylabel('월')

plt.show()
```

```python
df_uni = df_cate.groupby(['월', '카테고리'])['코스(상품) 이름'].nunique().reset_index(name='상품수')
```

```python
# 두 DataFrame 병합
df_merged = pd.merge(df_grouped, df_uni, on=['월', '카테고리'])
df_merged['강의 1개당 판매된 수'] = df_merged['count'] / df_merged['상품수']
```

```python
import matplotlib.pyplot as plt

# 피벗 테이블 생성
df_pivot = df_merged.pivot(index='월', columns='카테고리', values='강의 1개당 판매된 수')

# 그래프 크기 설정
plt.figure(figsize=(10, 6))

# 꺾은선 그래프 그리기
df_pivot.plot(kind='line')

# x축의 눈금을 1 단위로 설정
plt.xticks(range(min(df_pivot.index), max(df_pivot.index)+1))

plt.title('Monthly Sales Per Course')
plt.xlabel('Month')
plt.ylabel('Sales per Course')
plt.tight_layout() # 그래프의 레이블이 잘리는 것을 방지
plt.show()
```
