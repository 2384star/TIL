```python
import pandas as pd

# 월별로 DataFrame을 생성하고 CSV 파일로 저장하는 함수
def create_monthly_csv(df, month):
    # 해당 월의 데이터만 필터링
    monthly_df = df[df['월'] == month]

    # 일자별로 코스(상품) 이름의 카운트를 계산
    daily_counts = {}
    for day in range(1, 32):
        daily_df = monthly_df[monthly_df['일'] == day]
        counts = daily_df['코스(상품) 이름'].value_counts().to_dict()
        counts = {key: int(value) for key, value in counts.items()}  # 정수형으로 변환
        daily_counts[f'{day}일'] = counts

    # DataFrame으로 변환
    df_counts = pd.DataFrame(daily_counts).fillna(0)

    # CSV 파일로 저장
    csv_filename = f'{month}월 코스상품이름.csv'
    df_counts.to_csv(csv_filename, index_label='코스(상품) 이름')

# 각 월별로 CSV 파일 생성
for month in range(1, 13):
    create_monthly_csv(df, month)
```

```python
import pandas as pd

# 월별로 DataFrame을 생성하고 CSV 파일로 저장하는 함수
def create_monthly_csv(df, month):
    # 해당 월의 데이터만 필터링
    monthly_df = df[df['월'] == month]

    # 일자별로 코스(상품) 이름의 카운트를 계산
    daily_counts = {}
    for day in range(1, 32):
        daily_df = monthly_df[monthly_df['일'] == day]
        counts = daily_df['코스(상품) 이름'].value_counts().astype(int).to_frame().reset_index()
        counts.columns = ['코스(상품) 이름', day]
        counts.set_index('코스(상품) 이름', inplace=True)
        daily_counts[day] = counts

    # DataFrame으로 변환
    df_counts = pd.concat(daily_counts, axis=1).fillna(0).astype(int)

    # CSV 파일로 저장
    csv_filename = f'{month}월 코스상품이름.csv'
    df_counts.to_csv(csv_filename)

# 각 월별로 CSV 파일 생성
for month in range(1, 13):
    create_monthly_csv(df, month)
```

```python
import pandas as pd

# 중복 로우 제거
df = df.drop_duplicates()

# 공백칸이 있는 로우 제거
df = df.dropna(subset=['코스(상품) 이름'])

# 월별로 DataFrame을 생성하고 CSV 파일로 저장하는 함수
def create_monthly_csv(df, month):
    # 해당 월의 데이터만 필터링
    monthly_df = df[df['월'] == month]

    # 일자별로 코스(상품) 이름의 카운트를 계산
    daily_counts = {}
    for day in range(1, 32):
        daily_df = monthly_df[monthly_df['일'] == day]
        counts = daily_df['코스(상품) 이름'].value_counts().astype(int).to_frame().reset_index()
        counts.columns = ['코스(상품) 이름', day]
        counts.set_index('코스(상품) 이름', inplace=True)
        daily_counts[day] = counts

    # DataFrame으로 변환
    df_counts = pd.concat(daily_counts, axis=1).fillna(0).astype(int)

    # CSV 파일로 저장
    csv_filename = f'{month}월 코스상품이름.csv'
    df_counts.to_csv(csv_filename)

# 각 월별로 CSV 파일 생성
for month in range(1, 13):
    create_monthly_csv(df, month)
```
