## 차라리 하나하나 분리하자

```python
import matplotlib.pyplot as plt
import numpy as np

# 네오아카데미 코스
df_neo = df_re[df_re['코스(상품) 이름'].str.contains('네오아카데미')]
count_neo = df_neo['코스(상품) 이름'].value_counts(sort=False)

# 각 유니크한 값의 평균 실거래금액
average_price_neo = df_neo.groupby('코스(상품) 이름')['실거래금액'].mean().round(2)

# 인덱스 순서 정렬
sorted_index = sorted(count_neo.index)

# 그래프 그리기
fig, ax1 = plt.subplots(figsize=(12, 13))

# 막대 그래프 - 코스(상품) 이름 빈도수
bar1 = ax1.bar(sorted_index, count_neo[sorted_index], color='orange', alpha=0.6)
ax1.set_ylabel('코스(상품) 이름 빈도수')

# 각 막대 그래프 값 텍스트로 표시
for rect in bar1:
    height = rect.get_height()
    ax1.text(rect.get_x() + rect.get_width() / 2, height, str(height), ha='center', va='bottom')

# 오른쪽 y축 설정
ax2 = ax1.twinx()

# 꺾은선 그래프 - 각 유니크한 값의 평균 실거래금액
line1 = ax2.plot(range(len(average_price_neo)), average_price_neo[sorted_index], color='blue', marker='o', label='평균 실거래금액')
ax2.set_ylabel('평균 실거래금액')

# 각 꺾은선 그래프 값 텍스트로 표시
for i, price in enumerate(average_price_neo[sorted_index]):
    ax2.text(i, price, str(price), ha='center', va='bottom')

# 축과 레이블 설정
ax1.set_xlabel('코스(상품) 이름')
ax1.set_title('네오아카데미 코스(상품) 이름 빈도수와 평균 실거래금액')

# 범례 표시
lines = line1 + [bar1]
labels = [l.get_label() for l in lines]
plt.legend(lines, labels, loc='best')

# x축 글씨 세로로 표시
ax1.set_xticklabels(sorted_index, rotation='vertical')

plt.xticks(np.arange(len(sorted_index)), sorted_index)
ax2.set_ylim(50000, 650000)

plt.tight_layout()

plt.show()
```

## 김영하

```python
import matplotlib.pyplot as plt

# 김영하 코스
df_kim = df_re[df_re['코스(상품) 이름'].str.contains('김영하')]
count_kim = df_kim['코스(상품) 이름'].value_counts(sort=False)

# 김영하 코스 평균 실거래금액
average_price_kim = df_kim.groupby('코스(상품) 이름')['실거래금액'].mean().round(2)

# 그래프 그리기
fig, ax1 = plt.subplots(figsize=(10, 16))

# 막대 그래프 - 김영하 코스 빈도수
bar1 = ax1.bar(count_kim.index, count_kim, color='orange', alpha=0.6, label='김영하 코스')
ax1.set_ylabel('코스(상품) 이름 빈도수')

# 꺾은선 그래프 - 김영하 코스 평균 실거래금액
ax2 = ax1.twinx()
line1 = ax2.plot(range(len(average_price_kim)), average_price_kim.values, color='blue', marker='o', label='김영하 평균 실거래금액')

ax2.set_ylabel('평균 실거래금액')

# 각 막대 그래프 값 텍스트로 표시
for rect in bar1:
    height = rect.get_height()
    ax1.text(rect.get_x() + rect.get_width() / 2, height, str(height), ha='center', va='bottom')

# 각 꺾은선 그래프 값 텍스트로 표시
for i, price in enumerate(average_price_kim.values):
    ax2.text(i, price, str(price), ha='center', va='bottom')

# 축과 레이블 설정
ax1.set_xlabel('코스(상품) 이름')
ax1.set_title('김영하 코스 - 빈도수와 평균 실거래금액')

# 범례 표시
lines = line1 + [bar1]
labels = [l.get_label() for l in lines]
plt.legend(lines, labels, loc='best')

# x축 글씨 세로로 표시
ax1.set_xticklabels(count_kim.index, rotation='vertical')

plt.xticks(range(len(count_kim)), count_kim.index)

plt.tight_layout()
plt.show()
```

## 아래코드는 혹시나 쓸까봐 ㅎ

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'AppleGothic'

# '네오아카데미'를 포함하는 코스와 '김영하'를 포함하는 코스의 판매 빈도 및 평균 가격
courses_neo = df_neo['코스(상품) 이름'].unique()
courses_kim = df_kim['코스(상품) 이름'].unique()

counts_neo = df_neo.groupby('코스(상품) 이름').size()
counts_kim = df_kim.groupby('코스(상품) 이름').size()

avg_price_neo = df_neo.groupby('코스(상품) 이름')['실거래금액'].mean()
avg_price_kim = df_kim.groupby('코스(상품) 이름')['실거래금액'].mean()

fig, ax1 = plt.subplots(figsize=(10, 6))

# 막대 그래프 - 판매 빈도
bar1 = ax1.barh(courses_neo, counts_neo, color='yellow', alpha=0.6, label='네오아카데미 판매 빈도')
bar2 = ax1.barh(courses_kim, counts_kim, color='pink', alpha=0.6, label='김영하 판매 빈도')
ax1.set_xlabel('판매 빈도')

# 꺾은선 그래프 - 평균 가격
ax2 = ax1.twiny()
line1 = ax2.plot(avg_price_neo, courses_neo, color='b', marker='o', label='네오아카데미 평균 가격')
line2 = ax2.plot(avg_price_kim, courses_kim, color='g', marker='o', label='김영하 평균 가격')
ax2.set_xlabel('평균 가격')

# 축과 레이블 설정
ax1.set_ylabel('코스(상품) 이름')
ax1.set_title('코스(상품) 이름별 판매 빈도 및 평균 가격 비교')

# 범례 표시
lines = line1 + line2
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc='lower right')

plt.show()
```
