```python
import pandas as pd
from konlpy.tag import Mecab

# CSV 파일 불러오기
df = pd.read_csv('실습데이터.csv')
df
```

```python

# 형태소 분석을 위한 분석기 초기화
mecab = Mecab()

# 분석할 열 선택
target_col = '코스(상품) 이름'

# 새로운 열 생성하여 형태소 분석 결과 저장
df['{}_morph'.format(target_col)] = df[target_col].apply(lambda x: mecab.morphs(x))
```
